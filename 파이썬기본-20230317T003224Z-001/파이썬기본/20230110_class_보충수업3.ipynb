## 상속(Inheritance)

```python
class Parent:
  def __init__(self, name) :
    self.name = name
```

```python
class Child(Parent):
  pass
```

```python
c = Child()
```

```python
c = Child("sungwon")
```

```python
c.name
```

```python
class Child(Parent):
  def __init__(self, name, age):
    super().__init__(name)
    self.age = age
  pass
```

```python
c = Child("seo", 47)
```

```python
c.name
```

```python
c.age
```

```python
class Parent:
  def print_name(self) :
    print("Parent")
```

```python
class Parent:
  name = "Parent"   
  def print_name(self) :
    print("Parent")
```

```python
class Child(Parent):
  name = "Child"   # overriding 대상이 아님(속성이기 때문에)
  def print_name(self) : # overriding 대상
    print("Child")
```

```python
ch = Child()
```

```python
ch.print_name()
```

## Class Object Attrubue : Class object

```python
class Association:
  # default constructor
  # Not Define __init__
  def other_init(self, other, name):
    other.name = name
```

```python
class Client:
  def __init__(self, name, age):
    # self.name = name
    _name = Association()
    _name.other_init(self,name)
    self.age = age
  pass
```

```python
cl = Client("sungwon", 47)
```

```python
cl.name
```

## Overriding

### Method overriding


```python
class Animal:
  def say(self, message):
    return message
```

```python
class Dog(Animal):
  def say(self, message):
    print("This is dog:", message)
```

```python
class Bird(Animal):
  def say(self, message):
        print("This is Bird:", message)
```

```python
animal = Animal()
```

```python
animal.say("AAA")
```

```python
dog = Dog()
```

```python
dog.say("BBB")
```

```python
bird = Bird()
bird.say("CCC")
```

### Overloading

```python
class Method :
  def __init__(self, obj):
    self.obj = obj

  def _value_check(self, value):
    return value if isinstance(value, int) else int(value)    
  
  def __add__(self, value):
    value = self._value_check(value)
    self.obj = [x + value for x in self.obj]
    # return self.obj
    print("__add__", self.obj)

  def __sub__(self, value):
    value = self._value_check(value)
    self.obj = [x - value for x in self.obj]
    print("__sub__")

  def __mul__(self, value):
    value = self._value_check(value)
    self.obj = [x * value for x in self.obj]
    print("__mul__")

  def __truediv__(self, value):
    value = self._value_check(value)
    self.obj = [int(x/value) for x in self.obj] 
    print("__truediv__")
```

```python
m = Method([1,2,3,4])
```
