###리스트 Comprehension 방식 사용하기

리스트 컴프리헨션은 리스트를 만드는 간결한 방법을 제공한다. 흔한 용도는, 각 요소가 다른 시퀀스나 이터러블의 멤버들에 어떤 연산을 적용한 결과인 리스트를 만들거나, 어떤 조건을 만족하는 요소들로 구성된 서브 시퀀스를 만드는 것.

예를 들어, 제곱수의 리스트를 만들고 싶다고 가정하자, 다음과 같다:

```
squares = []
for x in range(10):
  squares.append(x**2)

print(x)
squares
```

이것은 x 라는 이름의 변수를 만들고 (또는 덮어쓰고) 루프가 종료된 후에도 남아있게 만든다는 것에 유의. 어떤 side-effect도 없이, 제곱수의 리스트를 이런 식으로 계산할 수 있다:

```
squares = [x**2 for x in range(10)]
squares
```

이것이 더 간결하고 읽기 쉽다.

리스트 컴프리헨션은 표현식과 그 뒤를 따르는 for 절과 여러 개의 for 나 if 절들을 감싸는 대괄호로 구성됨. 그 결과는 새 리스트인데, for 와 if 절의 문맥에서 표현식의 값을 구해서 만들어진다. 예를 들어, 이 리스트 컴프리헨션은 두 리스트의 요소들을 서로 같지 않은 것끼리 결합한다:

```
combs = []
for x in [1,2,3]:
  for y in [3,1,4]:
    if x != y:
      combs.append((x, y))
combs
```

```
combs = [(x, y) for x in [1,2,3] for y in [3,1,4] if x != y]
combs
```

```
vec = [-4, -2, 0, 2, 4]
# create a new list with the values doubled
print([x*2 for x in vec])

# filter the list to exclude negative numbers
print([x for x in vec if x >= 0])

# apply a function to all the elements
print([abs(x) for x in vec])

# call a method on each element
freshfruit = ['  banana', '  loganberry ', 'passion fruit  ']
print([weapon.strip() for weapon in freshfruit])

# create a list of 2-tuples like (number, square)
print([(x, x**2) for x in range(6)])


# flatten a list using a listcomp with two 'for'
vec = [[1,2,3], [4,5,6], [7,8,9]]
print([num for elem in vec for num in elem])

```

리스트 컴프리헨션은 복잡한 표현식과 중첩된 함수들을 포함할 수 있다:

```
from math import pi
[str(round(pi, i)) for i in range(1, 6)]
```

### 리스트 메소드(List Methods)


#### append()

* 리스트 요소 추가
</br>Add an item to the end of the list. Equivalent to `a[len(a):] = [x]`.

```
list1 = ['one', 'two', 'three']
list1.append('four')
list1.append([1, 2, 3])
print(list1)
list1[len(list1):] = ['five']
print(list1)
```

#### sort()

* 리스트 정렬

```
list1 = [2, 3, 4, 1]
list1.sort()
print(list1)

fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
fruits.sort()
print(fruits)
```

#### reverse()

* 리스트 요소 반전

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
fruits.reverse()
print(fruits)
```

#### index()

* 리스트의 요소 값에 대한 인덱스를 반환

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
#fruits.index('banana')
fruits.index('banana', 4)  # Find next banana starting a position 4
```

#### insert()

* 리스트 요소 삽입</br>Insert an item at a given position. The first argument is the index of the element before which to insert, so a.insert(0, x) inserts at the front of the list, and a.insert(len(a), x) is equivalent to a.append(x).

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
fruits.insert(0, 'grape')
print(fruits)
```

#### remove()

* 리스트 요소 제거

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
ret = fruits.remove('orange')
print(fruits)
print("ret=", ret)
print(type(ret))
```

#### del

* 리스트 요소 제거 연산
* 리스트 자체 삭제

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
del fruits[0]
print(fruits)
del fruits
print("fruits access")
#print(fruits) 
# 이후에 이름 fruits를 참조하는 것은 에러입니다
```

#### pop()

* 리스트 요소를 방출

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
ret1 = fruits.pop()
print(type(ret1))
print("ret1=", ret1)
print(fruits)

ret2 = fruits.pop()
print("ret2=", ret2)
print(fruits)


# li = [1, 2, 3, 4, 5]
# ret2 = li.pop()
# print(type(ret2))
# print(ret2)
# print(li)
```

#### count()

* 리스트 요소의 갯수 계산

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
print(fruits.count('apple'))
num = fruits.count('tangerine')
print(num)
print("******************************")
#if (num.type == Nonetype)
```

#### extend()

* 리스트 확장</br>Extend the list by appending all the items from the iterable. Equivalent to `a[len(a):] = iterable`.

```
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print(fruits)
fruits.extend(['grape', 'tangerine'])
print(fruits)
```

## 튜플(Tuple)

* 리스트와 유사하지만 변경 불가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* ‘(‘와 ‘)’를 사용하여 표현 가능

```
t = 12345, 54321, 'hello!' #튜플 패킹이라고 함
print(type(t))
print(t)

print(t[0])

# Tuples may be nested:
u = t, (1, 2, 3, 4, 5)
print(type(u))
print(u)

print(())
print((1, 2, 3))
print(('one', 'two', 'three'))
print((1, 2, 3, 'one', 'two', 'three'))
print((1, 2, 3, ('one', 'two', 'three')))

# but they can contain mutable objects:
v = ([1, 2, 3], [3, 2, 1])
print(v)
```

* range를 사용하여 튜플 만들기

* 튜플 = tuple(range(시작, 끝, 증가폭))

```
a = tuple(range(-4, 10, 2))
print(a)

b = [1, 2, 3]
c = tuple(b)
print(c)

d = list('Hello')
print(d)
e = tuple(d)
print(e)
```

### 튜플 인덱싱(Tuple Indexing)

* 튜플의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
tu = ('one', 'two', 'three')
print(tu)
print(tu[0])
print(tu[-0])
print(tu[2])
print(tu[-3])
```

위 예시에서 보듯이, 출력되는 튜플은 항상 괄호로 둘러싸인다, 그래서 중첩된 튜플이 올바르게 해석됨; 종종 괄호가 필요하기는 하지만 (튜플이 더 큰 표현식의 일부일 때), 둘러싼 괄호와 함께 또는 없이 입력될 수 있다. <font color="yellow">튜플의 개별 항목에 대입하는 것은 가능하지 않지만, 리스트 같은 가변 객체를 포함하는 튜플을 만들 수는 있다.</font>

튜플이 리스트처럼 보인다 하더라도, 이것들은 다른 상황에서 다른 목적으로 사용된다. 튜플은 불변 이고, 보통 이질적인 요소들의 시퀀스를 포함한다. 요소들은 언패킹 (이 섹션의 뒤에 나온다) 이나 인덱싱 (또는 네임드 튜플 의 경우는 어트리뷰트로도) 으로 액세스한다. 리스트는 가변 이고, 요소들은 보통 등질적이고 리스트에 대한 이터레이션으로 액세스 된다.

특별한 문제는 비었거나 하나의 항목을 갖는 튜플을 만드는 것이다: 이 경우를 수용하기 위해 문법은 추가적인 예외 사항을 갖고 있다. 빈 튜플은 빈 괄호 쌍으로 만들어짐; `하나의 항목으로 구성된 튜플은 값 뒤에 쉼표를 붙여서 만든다` (값 하나를 괄호로 둘러싸기만 하는 것으로는 충분하지 않음) 예를 들어:

```
empty = ()
singleton = ('hello')
#singleton = 'hello',    # <-- note trailing comma
print(type(singleton))
print(len(empty))
print(len(singleton))
print(singleton)
```

문장 t = 12345, 54321, 'hello!' 는 튜플 패킹 의 예:</br> 값 12345, 54321, 'hello!' 는 함께 튜플로 패킹 됨. 반대 연산 또한 가능함:

```
print(t)
x, y, z = t # t = 12345, 54321, 'hello!' 
print(x)
print(y)
print(z)
print(x, y, z)
```

이것을 언패킹이라고 함.튜플 언패킹을 할 때 튜플의 일부 값을 하나의 변수로 묶어서 바인딩할 수도 있음. 예를 들어 올림픽 체조 경기의 점수 합산 방식처럼 가장 낮은 값과 가장 높은 값은 각각 low, high로 바인딩하고 나머지 값들은 others라는 변수로 바인딩하려면 다음과 같이 변수 앞에 *를 하나 붙여주면 된다.

[참조](https://wikidocs.net/92538)

```
scores = (1, 2, 3, 4, 5, 6)
low, *others, high = scores
print(others)
```

이것은 시퀀스 언패킹 이라고 불리고 오른쪽에 어떤 시퀀스가 와도 됨. 시퀀스 언 패킹은 등호의 좌변에 시퀀스에 있는 요소들과 같은 개수의 변수들이 올 것을 요구함. 다중 대입은 사실 튜플 패킹과 시퀀스 언 패킹의 조합일뿐이라는 것에 유의해야함.

```
li = [1,2,3,4,5,6]
low, *others, high = li
print(others)
```

### 언패킹과 함수 호출

아래 코드처럼 4개의 수를 입력 받아 모두 더한 값을 리턴하는 hap 함수가 있다. 이 함수를 사용하려면 4개의 데이터를 다음과 같이 전달해줘야 함.

```
def hap(num1, num2, num3, num4):
    return num1+num2+num3+num4

scores = (1, 2, 3, 4)
result = hap(scores[0], scores[1], scores[2], scores[3])
print(result)
```

위와 같은 함수 호출코드는 다음과 같이 튜플 언패킹 기능을 통해 간소화할 수 있다.

```
def hap(num1, num2, num3, num4):
    return num1+num2+num3+num4

scores = (1, 2, 3, 4)
result = hap(*scores)            # 함수 호출 시 튜플 언패킹
print(result)
```

## 세트(Set)

* 데이터 중복을 허용하지 않는 구조
* 순서가 없는 데이터 집합을 위한 구조
* 인덱싱으로 값을 접근할 수 없음
* ‘{‘와 ‘}’를 사용하여 표현

집합을 만들 때는 중괄호나 set() 함수를 사용할 수 있음. 주의사항: 빈 집합을 만들려면 set() 을 사용해야 한다. {} 가 아님; 후자는 빈 딕셔너리를 만드는데, 다음 섹션에서 확인함.

```
emptyset = {}
print(type(emptyset))

emptyset = set()
print(type(emptyset))

print({10, 20, 30, 40})
print({'one','two', 'three', 'three', 'one'}) # 데이터 중복을 허용하지 않는 구조
```

### 세트 연산자(Set Operators)

* 교집합: &
* 합집합: |
* 차집합: -
* 여집합: ^

```
set1 = {10, 20, 20, 30}
set2 = {30, 30, 40, 50}
print(set1)
print(set2)
print("교집합:", set1 & set2)
print("합집합:", set1 | set2)
print("차집합:", set1 - set2)
print("여집합:", set1 ^ set2)
```

### 세트 메소드(Set Methods)

* 교집합: intersection()
* 합집합: union()
* 차집합: difference()
* 여집합: symmetric_difference()

```
set1 = {10, 20, 20, 30}
set2 = {30, 30, 40, 50}
print(set1)
print(set2)
print(set1.intersection(set2))
print(set1.union(set2))
print(set1.difference(set2))
print(set1.symmetric_difference(set2))
```

* 요소 추가: add()
* 여러 요소 추가: update()
* 요소 제거: remove()
* 요소 제거: discard()
* 모든 요소 제거: clear()

```
set1 = {10, 20, 20, 30}
print(set1)

set1.add(50)
print(set1)

set1.update((60, 70)) #set1.update([60, 70])
print(set1)

ret1 = set1.remove(60)
print(set1)
print(ret1)

ret2 = set1.discard(30)
print(set1)
print(ret2)

set1.clear()
print(set1)
```

## 딕셔너리(Dictionary)

* 키(key)와 값(value)의 쌍으로 구성된 데이터
* 순서가 없는 데이터
* 키를 통해 값을 얻음
* 동일한 키가 있을경우 덮어씀

```
#dic1 = {1:'one', 2:'two', 3:'three'}
#dic1 = dict([(1, 'one'), (2, 'two'), (3, 'three')])
#dic1 = dict(1=one, 2=two, 3=three) 안됨
#키가 간단한 문자열일 때, 때로 키워드 인자들을 사용해서 쌍을 지정하기가 쉬움:
dic1 = dict(one=1, two=2, three=3)

print(dic1)
print(type(dic1))

dic2 = {1:'one', 2:'two', 3:'thre', 1:'one', 2:'two', 3:'three'}
print(dic2)

dic3 = {1:'one', 2:'two', 3:'thre', 1:'1', 2:'2', 3:'3'}
print(dic3)

dic4 = {1:'one', 2:'two', 3:'thre', 1:'1', 2:'2', 3:'3', 1:'first', 2:'second', 3:'third'}
print(dic4)

#print(dic1[1])
print(dic2[2])
print(dic3[3])

print({x: x**2 for x in (2, 4, 6)})

name = ['merona', 'gugucon']
price = [500, 1000]
icecream = dict(zip(name, price))
print(icecream)

icecream = {k:v for k, v in zip(name, price)}
print(icecream)

icecream2 = {k:v*2 for k, v in zip(name, price)}
print(icecream2)

name = ['merona', 'gugucon', 'bibibig']
price = [500, 1000, 600]

icecream = {k:v for k, v in zip(name, price) if v < 1000}
print(icecream)
```

* 숫자들로 인덱싱되는 시퀀스와 달리, 딕셔너리는 키 로 인덱싱되는데, 모든 불변형을 사용할 수 있다; 문자열과 숫자들은 항상 키가 될 수 있다. 
* 튜플이 문자열, 숫자, 튜플들만 포함하면, 키로 사용될 수 있다; 튜플이 직접적이나 간접적으로 가변 객체를 포함하면, 키로 사용될 수 없다. 
* 리스트는 키로 사용할 수 없는데, 리스트는 인덱스 대입, 슬라이스 대입, append() 나 extend() 같은 메서드들로 값이 수정될 수 있기 때문임

```
tel = {'jack': 4098, 'sape': 4139}
```

### 딕셔너리 요소 추가/삭제

* 딕셔너리의 해당 키 값에 값을 추가하여 요소 추가
* del을 이용하여 요소 제거

```
dic1 = {1:'one', 2:'two', 3:'thre'} #dic1 = dict()
print(dic1)

print(dic1[2])
dic1[4] = 'four'
print(dic1)

dic1[5] = 'five'
print(dic1)

tel = {'jack': 4098, 'sape': 4139}
tel['guido'] = 4127
print(tel)

print('guido' in tel) #하나의 키가 딕셔너리에 있는지 검사하려면, in 키워드들 사용
print('jack' not in tel)

print(list(tel)) #딕셔러리에 list(d) 를 수행하면 딕셔너리에서 사용되고 있는 모든 키의 리스트를 삽입 순서대로 리턴함

del dic1[4]
print(dic1)

```

### 딕셔너리 메소드(Dictionary Methods)



|메소드|설명|
|--|--------|
|keys()|딕셔너리의 키 가져오기|
|values()|딕셔너리의 값 가져오기|
|items()|딕셔너리의 키와 값을 모두 가져오기|
|get()|딕셔너리에서 키에 해당하는 값 가져오기|
|pop()|딕셔너리에서 키에 해당하는 값 추출하기|
|clear()|딕셔너리의 모든 요소를 제거하기|

```
dic1 = {1:'one', 2:'two', 3:'three'} 
print(dic1)
print(dic1.keys())
print(dic1.values())
print(dic1.items())
print(dic1.get(1))
print(dic1.pop(3))
dic1.clear()
print(dic1)
```
