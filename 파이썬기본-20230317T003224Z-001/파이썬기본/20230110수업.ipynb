```python
class Klass:
#  생성자는 안만들어도 된다.아래와 같이 __new__ 생성자를 정의하지 않으면 인스턴스 만들때 디폴트 생성자가 자동으로 호출된다. 
  def __new__(cls, name, age): #생성자 : Constructor , cls:Klass라는 클래스를 나타냄
    print("객체가 생성됨")
    return object.__new__(cls) #object는 최상위클래스

  def __init__(self, name, age): #__init__는 생성자함수가 아니다. 초기화 함수이다. 
    self.name = name #인스턴스의 속성
    self.age = age #인스턴스의 속성
    print("객체가 초기화됨")

  def __del__(self): #소멸자(Destructor), self가 있으므로 객체 자기 자신을 지운다는 뜻.
    print("객체가 소멸됨")
    del self
```

Static method

```python
class SMethod: 
  _name = "Class Attirbute" 
  def __init__(self, name):
    self._name = name
  
  @staticmethod   
  def name(cls) :  
    return cls._name 

  @staticmethod   
  def name(self) :  
    return self._name

  @staticmethod   
  def func(x,y) : #빌트인 데이터타입이 주로 static method지원해준다.
    return x+y
   
```

##Name mangling



```python
class Naming__:
  def __init__(self, name, age):
    self._name = name
    self._age = age

  def __getName(self): #__를 붙이면 외부에서 호출이 안된다. 클래스 내부에서만 사용가능, 언더바를 하나쓰면 관습적으로 외부에서 함수를 호출하지 말라는 뜻임.
    return self._name
  
  def __getAge(self):  
    return self._age
  
  def getAttr(self):
    return self.__getName(), self.__getAge() #두 함수를 클래스 내부에서만 사용 중

  
```

```python
n = Naming__("seo", 47)
```

```python
n.__getName()  #함수이름에 __가 붙어있어서 외부에서 호출되지 않는다.
```

```python
n.getAttr()
```

```python
class Property:
  def __init__(self, name, age):
    self._name = name
    self._age = age

  @property     #getter역할
  def name(self): 
    return self._name
  
  @name.setter #setter역할
  def name(self, name):  
    self._name = name
  
  @name.deleter
  def name(self):
    del self._name 
```

##상속(Inheritance)

```python
class Parent:
  def __init__(self, name):
    self.name = name
```

```python
class Child(Parent): #Parent class를 상속했다.(부모클래스 = 베이스클래스 = 최상위클래스)
  pass              #아무것도 작성하지 않을 시
```

```python
c = Child() #parent를 상속했으므로 인수를 넣어줘야 한다.
```

```python
c.name
```

```python
class Child(Parent):
  def __init__(self, name, age): 
    super().__init__(name)       #sup모의 naer는 Parent, 부me설정
    self.age = age
  pass
```

```python
c = Child("seo", 47)
```

## Class Object Attribute : Class object 

```python
class Association:
  #default constructor 
  #Not Define __init__
  def other_init(self, other, name):
    other.name = name
```

```python
class Client:
  def __init__(self, name, age): 
    #self.name = name
    _name = Association()           #init함수안에서 특정 클래스 객체를 만든다.
    _name.other_init(self,name)     #객체 메서드 호출
    self.age = age
  pass
```

```python
cl = Client("sungwon", 47)
```

```python
cl.name
```

##오버로딩 Overloading

```python
class Method :
  def __init__(self, obj):
    self.obj = obj

  def _value_check(self, value):
    return value if isinstance(value, int) else int(value)    
  
  def __add__(self, value):
    value = self._value_check(value)
    self.obj = [x + value for x in self.obj]
    print("__add__")

  def __sub__(self, value):
    value = self._value_check(value)
    self.obj = [x - value for x in self.obj]
    print("__sub__")

  def __mul__(self, value):
    value = self._value_check(value)
    self.obj = [x * value for x in self.obj]
    print("__mul__")

  def __truediv__(self, value):
    value = self._value_check(value)
    self.obj = [int(x/value) for x in self.obj]
    print("__truediv__")
```

```python
m = Method([1,2,3,4])
```

```python
m = m + 3 #'+'연산자 오버로딩, +연산자가 클래스 내 특정 함수로 오버로딩 되어있다. +연산의 기능이 인스턴스연산이 가능하도록 변형되어 있음.
```
