### 사각형 그리기
- cv2.rectangle(img_src, start, end, color, thickness)
- img_src  :  그림을 그릴 이미지 파일
- start  : 시작 좌표 pt1(x1, y1) 
- end  : 종료 좌표 pt2(x2, y2)
- color  : BGR형태의 Color( (B, G, R) : (255, 0, 0))
- thickness (int) – 선의 두께(pixel), -1이면 안쪽을 채움

```python
import cv2
import numpy as np

height = 600
width = 800
img_zero = np.zeros((height,width,3), dtype=np.uint8)
green = (0,255,0)
cv2.line(img_zero, (0,0),(width-1,height-1), green, 1)
cv2.line(img_zero, (width-1,0),(0,height-1), green, 1)

red = (0,0,255)
pt1 = (int(width/4),int(height/4))
pt2 = (int(3*width/4),int(3*height/4))
cv2.rectangle(img_zero, pt1,pt2, red, 2)

#cv2.rectangle(img_src, pt1,pt2, color, thickness=-1) : thickness -1이면 채움
pt3 = (int(7*width/16),int(7*height/16))
pt4 = (int(9*width/16),int(9*height/16))
cv2.rectangle(img_zero, pt3,pt4, green, -1)

cv2.imshow('src',img_zero)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# matplotlib에서 영상 띄우기
show_matplotlib_image(img_zero, 7,6)
```

### 타원
- cv2.ellipse(img_src, center, axes, angle, startAngle, endAngle, color, thickness)

- img_src  :  그림을 그릴 이미지 파일
- center  :  타원의 중심좌표 (x, y) 
- axes  :  타원의 가장 큰 거리와 작은 거리
- angle :  타원의 회전 각도(시계방향)
- startAngle : 타원의 시작 각도
- endAngle  : 타원이 끝나는 각도
- color  : 타원의 색 BGR형태( (B, G, R) : (255, 0, 0))
- thickness (int) – 선의 두께(pixel), -1이면 안쪽을 채움

```python
import cv2
import numpy as np

height = 600
width = 800
img_zero = np.zeros((height,width,3), dtype=np.uint8)

green = (0,255,0)
cv2.line(img_zero, (0,0),(width-1,height-1), green, 1)
cv2.line(img_zero, (width-1,0),(0,height-1), green, 1)

red = (0,0,255)
pt1 = (int(width/4),int(height/4))
pt2 = (int(3*width/4),int(3*height/4))
cv2.rectangle(img_zero, pt1,pt2, red, 2)

#cv2.rectangle(img_src, pt1,pt2, color, thickness=-1) : thickness -1이면 채움
pt3 = (int(7*width/16),int(7*height/16))
pt4 = (int(9*width/16),int(9*height/16))
cv2.rectangle(img_zero, pt3,pt4, green, -1)

# cv2.circle(img,center,radian,color,thickness)
cv2.circle(img_zero,(int(width/2),int(height/2)), 250, (255,255,0), 1)
cv2.circle(img_zero,(int(width/2),int(height/2)), 63, (255,0,255), 1)

# cv2.ellipse(img_src, center, axes, angle, startAngle, endAngle, color, thickness)
cv2.ellipse(img_zero, (int(width/4),int(height/4)), (100,50), 0, 90, 360, (0,255,0), -1)

cv2.imshow('src',img_zero)
cv2.waitKey()
cv2.destroyAllWindows()
```

### 영상에 한글 쓰기
- PIL : python image library를 통해 우회

```python
import cv2
import numpy as np
from PIL import ImageFont, ImageDraw, Image

def input_korean(img_dst, input_text, pt, color, fontsize=20):
    my_font = ImageFont.truetype("font/gulim.ttc", fontsize)
    img_dst = Image.fromarray(img_dst)
    mydraw = ImageDraw.Draw(img_dst)
    mydraw.text((pt[0], pt[1]), input_text, font=my_font, fill=(color[0], color[1], color[2]))
    return np.array(img_dst)

height = 600
width = 800
img_zero = np.zeros((height,width,3), dtype=np.uint8)

### 한글 추가 부분 ####################################
pt = (100,100)
color = (0,255,255)
my_text = "안녕하세요"
img_zero = input_korean(img_zero, my_text, pt, color)
###########################################################

cv2.putText(img_zero, 'Hello', (int(width/2)-37,int(height/2)+10),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

cv2.imshow('src',img_zero)
cv2.waitKey()
cv2.destroyAllWindows()
```
