##동영상출력과 동영상 자르기

```python
import cv2

# 입력 및 출력 파일 경로 정의
input_video_path = '/content/drive/MyDrive/eye_blink_detector/eye_blink_detector/videos/sleep_videoplay.mp4'
output_video_path = 'output.mp4'

# 입력 동영상 파일 열기
capture = cv2.VideoCapture(input_video_path)

# 영상의 초당 프레임 수(FPS) 및 너비와 높이 구하기
fps = capture.get(cv2.CAP_PROP_FPS)
frame_width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 출력 동영상 파일을 위한 VideoWriter 객체 생성
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 코덱 정의
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# 10초에 해당하는 프레임 수 계산
frames_to_skip = int(10 * fps)

# 해당 프레임 수만큼 이동
capture.set(cv2.CAP_PROP_POS_FRAMES, frames_to_skip)

while True:
    # 프레임 읽기
    ret, frame = capture.read()
    if not ret:
        # 프레임이 더 이상 없으면 종료
        break

    # 프레임 쓰기 (출력 동영상 파일에 추가)
    out.write(frame)

# 동영상 파일 닫기
capture.release()
out.release()
```

##구글드라이브에 동영상 잘라서 바로저장

```python
import cv2

input_video_path = '/content/drive/MyDrive/eye_blink_detector/eye_blink_detector/videos/sleep_eye.mp4'
# 동영상 파일 불러오기
cap = cv2.VideoCapture(input_video_path)

# 동영상 파일의 프레임 수, FPS, 너비, 높이 가져오기
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 뒤에서 10초를 제외한 프레임 수 계산
frames_to_keep = total_frames - int(20 * fps)

# 출력 동영상 파일 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/drive/MyDrive/output.mp4', fourcc, fps, (width, height))

for i in range(frames_to_keep):
    ret, frame = cap.read()
    if ret:
        out.write(frame)

cap.release()
out.release()
```

## 동영상 절반만 자르기

```python
import cv2

# 동영상 파일 경로
input_video_path = '/content/drive/MyDrive/eye_blink_detector/eye_blink_detector/videos/drive.mp4'
# 동영상 파일 불러오기
cap = cv2.VideoCapture(input_video_path)

# 동영상 파일의 FPS, 너비, 높이, 전체 프레임 수 가져오기
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# 오른쪽 반만 잘라내기 위한 너비 계산
width_crop = width // 2

# 출력 동영상 파일 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/drive/MyDrive/output.mp4', fourcc, fps, (width_crop, height))

# 동영상의 모든 프레임을 순회하며 오른쪽 부분 잘라내기
for i in range(total_frames):
    ret, frame = cap.read()
    if ret:
        frame_crop = frame[:, width_crop:]
        out.write(frame_crop)

cap.release()
out.release()
```
