```python
data  = [(167,71),(182,69),(173,60),(163,45)]

a = 0.94
b=-100
def func(x):
  return a *x + b

total = 0
for x, y in data:
  pred = func(x)
  error = pred -y
  print(x, y, pred, error)
  total += (pred - y) ** 2

total
```
