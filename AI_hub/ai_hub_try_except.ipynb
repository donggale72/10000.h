```python
def print_page(page):
  for text in page['text']:
    print(text)

  for index, select in enumerate(page['select']):
    print(index + 1, select[0])

print_page(data['a'])
```

```python
page = data['a']
i=0
while i <= 5:
  print_page(page)
  index = int(input('?'))-1
  next_page_name =page['select'][index][1]
  page = data[next_page_name]
  i+=1

print('마침니다.')
```
