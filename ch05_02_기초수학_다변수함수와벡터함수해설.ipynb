### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x):
  return x**2+2*x+1
f1(3)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10,10,100)
fs = f1(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, fs, alpha=.8)
ax.set_title('$f(x)$ according to $x$')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')

x = -1
plt.plot(x, f1(x), marker='o', c='r')
plt.text(x, f1(x), f'({x}, {np.round(f1(x),2)})')
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상의 모든 실수
  * 그래프의 특징 : x값이 -1일때 아웃풋 값이 가장 작으며, 양쪽으로 아웃풋 값은 커진다

#### **예제 02**

\begin{align}
f(x) = cosin(x)
\end{align}

* 함수 생성하기

````python
def f2(x):
  return np.cos(x)
f2(5)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10,10,100)
fs = f2(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : -1과 1 사이 모든 실수
  * 그래프의 특징 :주기적으로 -1과 1 사이를 횡보한다.

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x, y):
  return x+y 
f3(1,1)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15,20,20)
ys = np.linspace(-15,20,20)
X, Y = np.meshgrid(xs,ys)
F = f3(X, Y)
X.shape, Y.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d')
ax.scatter3D(X, Y, F, c='k', marker='.')
surf = ax.plot_surface(X, Y, F, cmap='rainbow', alpha=.7)

ax.set_title('$f(x,y)$ acording to x, y')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
fig.colorbar(surf, shrink=.6, label='f(x,y)')
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 모든실수
  * 그래프의 특징 :x, y 값이 함께 증가하는 방향으로 아웃풋 f(x) 도 커진다

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x, y):
  return (x-5)**2+(y-5)**2 
f4(2,2)
````

* 인풋 정의(x, y 모두 -15~20 사이 값)

````python
xs = np.linspace(-15,20,20)
ys = np.linspace(-15,20,20)
X, Y = np.meshgrid(xs,ys)
F = f4(X,Y)
X.shape, Y.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
ax.scatter3D(X, Y, F, c='k', marker='.')
surf = ax.plot_surface(X, Y, F, cmap='rainbow', alpha=.5)

ax.set_title('f(x,y) acording to x, y')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
fig.colorbar(surf, shrink=.6, label='f(x,y)')
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 0보다 큰 모든 실수
  * 그래프의 특징 : x=5, y=5 일때 최소값을 가지고 이 값에서 멀어질수록 아웃풋 f(x) 도 커진다

#### **예제 05**

\begin{align}
f(x_1, x_2) = 50(x_2-x_1^2)^2+(2-x_1)^2
\end{align}

* 함수 생성하기

````python
def f5(x1, x2):
  return 50*(x2-(x1**2))**2 + (2-x1)**2
f5(2,4)
````

* 인풋 정의 (x: -2 ~ 2 20개, y: -1~3 20개)

````python
xs1 = np.linspace(-2, 2, 20)
xs2 = np.linspace(-1, 3, 20)
X1, X2 = np.meshgrid(xs1,xs2)
F = f5(X1, X2)
X1.shape, X2.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
ax.scatter3D(X1, X2, F, c='k', marker='.')
surf = ax.plot_surface(X1, X2, F, cmap='rainbow', alpha=.5)

ax.set_title('$f(x_1, x_2)$ acording to $x_1$ and $x_2$')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1,x_2)$')
fig.colorbar(surf, shrink=.6, label='$f(x_1, x_2)$')
plt.show()
````

* $x_1$, $x_2$의 범위를 둘다 -4부터 4로 변경하여 다시 그려보세요.

````python
xs1 = np.linspace(-4, 4, 20)
xs2 = np.linspace(-4, 4, 20)
X1, X2 = np.meshgrid(xs1,xs2)
F = f5(X1, X2)
X1.shape, X2.shape, F.shape

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
ax.scatter3D(X1, X2, F, c='k', marker='.')
surf = ax.plot_surface(X1, X2, F, cmap='rainbow', alpha=.5)

ax.set_title('$f(x_1, x_2)$ acording to $x_1$ and $x_2$')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1,x_2)$')
fig.colorbar(surf, shrink=.6, label='$f(x_1, x_2)$')
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 0보다 큰 모든 실수
  * 그래프의 특징 : x1=2, x2=4 일때 최소값 0을 가진다. x2의 변화보다 x1의 변화가 아웃풋 값에 큰 영향을 미친다. (x1의 제곱 함수기 때문에)

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
  g = 5*x
  h = x**2
  return   np.array([g, h])

f6(5)
````

* 인풋 정의 (-10~10 20개)

````python
xs = np.linspace(-10,10,20)
gs, hs = f6(xs)
xs.shape, gs.shape, hs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)

ax.plot(gs, hs)
ax.set_title('$f(x)$ space')
ax.set_xlabel('$g(x)$')
ax.set_ylabel('$h(x)$')

scat = ax.scatter(gs, hs, c=xs, cmap='rainbow')
fig.colorbar(scat)
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든 실수
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 0보다 큰 모든 실수

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
  x = np.sin(6*t)
  y = 1/4*t
  z = t**2/2
  np.array([x, y, z])
  return   np.array([x, y, z])
f7(3)
````

* 인풋 정의 (-1~3 200개)

````python
ts = np.linspace(-1,3,200)
xs, ys, zs = f7(ts)

ts.shape, xs.shape, ys.shape, zs.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
scat = ax.scatter3D(xs, ys, zs, c=ts, marker='o', cmap='rainbow')

ax.set_title('$f(t)$ space')
ax.set_xlabel('$x(t)$')
ax.set_ylabel('$y(t)$')
ax.set_zlabel('$z(t)$')
fig.colorbar(scat, shrink=.6, label='t')
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1과 1 사이 수
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 모든 실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 0보다 큰 모든 실수

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f8(t):
  return np.array([np.cos(t), np.sin(t), t])
f8(3)
````

* 인풋 정의(-10 ~ 10 사이 200개)

````python
ts = np.linspace(-10,10,200)
xs, ys, zs = f8(ts)

ts.shape, xs.shape, ys.shape, zs.shape
````

````python
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
scat = ax.scatter3D(xs, ys, zs, c=ts, marker='.', cmap='rainbow')

ax.set_title('$f(t)$ space')
ax.set_xlabel('$cos(t)$')
ax.set_ylabel('$sin(t)$')
ax.set_zlabel('$t$')
fig.colorbar(scat, shrink=.6, label='t')
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * cos(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1과 1 사이 수
  * sing(t)가 가질 수 있는 값 (두번째 아웃풋) : -1과 1사이 수


## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x, y):
  return np.array([x**2, y**2, x*y])
f9(1, 2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-.5,1,50)
ys = np.linspace(-.5,1,50)
X, Y = np.meshgrid(xs, ys)
Fs = f9(X, Y)
X.shape, Y.shape, Fs.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(10,7))

ax = plt.axes(projection='3d')
ax.scatter3D(Fs[0], Fs[1], Fs[2], marker='.')

ax.set_title('$f(x,y)$ space')
ax.set_xlabel('$x^2$')
ax.set_ylabel('$y^2$')
ax.set_zlabel('$xy$')
plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u, v):
  return np.array([u, v, 1+u**2+v/(1+v**2)])
f10(1,2)
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
us = np.linspace(0,1,50)
vs = np.linspace(0,1,50)
U, V = np.meshgrid(us, vs)
F = f10(U, V)
U.shape, V.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,9))

ax = plt.axes(projection='3d')
ax.scatter3D(F[0], F[1], F[2], marker='.')

ax.set_title('$f(u,v)$ space')
ax.set_xlabel('$u$')
ax.set_ylabel('$v$')
ax.set_zlabel('$1+u^2+v/(1+v^2)$')
plt.show()
````
