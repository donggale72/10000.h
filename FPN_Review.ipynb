# Abstract
Feature pyramids are a basic component in recognition systems for detecting objects at different scales. But recent deep learning object detectors have avoided pyramid rep- resentations, in part because they are compute and memory intensive. In this paper, we exploit the inherent multi-scale, pyramidal hierarchy of deep convolutional networks to con- struct feature pyramids with marginal extra cost. A top- down architecture with lateral connections is developed for building high-level semantic feature maps at all scales. This architecture, called a Feature Pyramid Network (FPN), shows significant improvement as a generic feature extrac- tor in several applications. Using FPN in a basic Faster R-CNN system, our method achieves state-of-the-art single- model results on the COCO detection benchmark without bells and whistles, surpassing all existing single-model en- tries including those from the COCO 2016 challenge win- ners. In addition, our method can run at 6 FPS on a GPU and thus is a practical and accurate solution to multi-scale object detection. Code will be made publicly available.


Feature Pyramid는 다양한 스케일에서 물체를 감지하기 위한 인식 시스템의 기본 구성 요소입니다. 그러나 최근의 딥러닝 객체 검출기들은 계산 및 메모리 부담이 크다는 이유로 피라미드 representations을 피하고 있습니다. 이 논문에서는 딥 컨볼루션 네트워크의 다중 스케일, 피라미드 형태의 계층 구조를 활용하여 더 적은 비용으로 특징 피라미드를 구축합니다. top-down 구조의 lateral connection을 사용하여 모든 스케일에서 high-level semantic feature maps을 구성하는 FPN(Feature Pyramid Network) 아키텍처를 개발했습니다. 이 아키텍처는 여러 응용 분야에서 일반적인 feature extractor로서 유의미한 성능 향상을 보여줍니다. FPN을 Faster R-CNN 시스템에 적용하여 COCO detection 벤치마크에서 모든 기능을 사용하지 않은 단일 모델 결과로 최고 성능을 달성했습니다. 이는 COCO 2016 챌린지 우승자들의 단일 모델 엔트리를 포함하여 모든 기존 단일 모델 엔트리를 뛰어넘습니다. 또한, 우리의 방법은 GPU에서 6 FPS로 실행되므로 multi scale 객체 검출의 실용적이고 정확한 솔루션입니다. 코드는 공개적으로 사용 가능할 예정입니다.

built upon image pyramids (for short we call these featurized image pyramids) form the basis of a standard solution

## scale-invariant

* 스케일 불변성은 명시적인 스케일 조정 작업없이 서로 다른 스케일에서도 객체 또는 패턴을 인식할 수 있는 알고리즘 또는 시스템의 특성을 말합니다. 즉, 스케일 불변성 시스템은 크기나 확대/축소에 관계없이 동일한 객체 또는 패턴을 감지할 수 있습니다.

* 컴퓨터 비전에서 스케일 불변성은 시야, 거리 및 방향과 같은 요소로 인해 이미지의 객체가 서로 다른 스케일로 나타날 수 있기 때문에 중요합니다. 예를 들어, 카메라에서 멀리 떨어진 차는 카메라에 가까운 동일한 차보다 작게 보입니다. 따라서 스케일 불변성 알고리즘은 각각의 스케일에 대해 별도로 처리하지 않고 두 상황 모두에서 차를 감지할 수 있습니다.

* 특징 피라미드는 컴퓨터 비전에서 스케일 불변성을 달성하기 위해 널리 사용되는 기술입니다. 서로 다른 스케일로 이미지 피라미드를 구축하고 각 레벨에서 특징을 계산함으로써 특징 피라미드는 서로 다른 스케일의 객체를 감지할 수 있습니다. 이는 엣지와 코너와 같은 객체 인식에 중요한 특징이 여러 스케일에서 발생하기 때문입니다.

* 종합적으로, 스케일 불변성은 객체의 스케일이 크게 다양한 실제 상황에서 객체를 인식할 수 있는 능력을 제공하여 시스템이 더 견고하고 정확한 인식 능력을 갖도록 합니다.

Featurized image pyramids were heavily used in the era of hand-engineered features [5, 25]. They were so
critical that object detectors like DPM [7] required dense
scale sampling to achieve good results (e.g., 10 scales per
octave). For recognition tasks, engineered features have largely been replaced with features computed by deep convolutional networks (ConvNets) [19, 20]. Aside from being
capable of representing higher-level semantics, ConvNets
are also more robust to variance in scale and thus facilitate
recognition from features computed on a single input scale
[15, 11, 29] (Fig. 1(b)). But even with this robustness, pyramids are still needed to get the most accurate results. All recent top entries in the ImageNet [33] and COCO [21] detection challenges use multi-scale testing on featurized image
pyramids (e.g., [16, 35]). The principle advantage of featurizing each level of an image pyramid is that it produces
a multi-scale feature representation in which all levels are
semantically strong, including the high-resolution levels.

Featurized 이미지 피라미드는 핸드 엔지니어링된 기능의 시대에서 광범위하게 사용되었습니다. 이러한 기능들은 객체 검출기(DPM)와 같은 객체 검출기에서 좋은 결과를 얻으려면 밀도가 높은 스케일 샘플링(예: 옥타브 당 10개의 스케일)이 필요한 정도로 중요합니다. 인식 작업에서는 엔지니어링된 기능 대신 딥 컨볼루션 신경망(ConvNets)으로 계산된 기능이 크게 대체되었습니다. 고수준 의미를 나타내는 능력 외에도 ConvNets는 스케일 변화에 대한 강인성이 더 높기 때문에 단일 입력 스케일에서 계산된 기능으로부터 인식을 촉진합니다. 하지만 이러한 강인성이 있더라도 여전히 가장 정확한 결과를 얻기 위해서는 피라미드가 필요합니다. ImageNet 및 COCO 검출 챌린지의 최근 상위 항목들은 모두 featurized 이미지 피라미드에서 멀티스케일 테스트를 사용합니다. 이미지 피라미드의 각 수준을 기능화하는 원칙적인 이점은 모든 수준이 고해상도 수준을 포함하여 모든 수준이 의미론적으로 강력한 멀티스케일 기능 표현을 생성한다는 것입니다.

## hand-engineered features
* Hand-engineered features는 문제 도메인을 이해한 인간이 디자인하고 계산하는 feature를 의미합니다. 컴퓨터 비전 분야에서는 딥 러닝 기반 방법이 등장하기 전에 광범위하게 사용되었습니다.

* 객체 감지에서 hand-engineered feature는 일련의 사전 정의된 규칙과 휴리스틱을 기반으로 계산되며, 객체의 본질적인 특성을 포착하기 위한 것입니다. 예를 들어 HOG(Histogram of Oriented Gradients) feature를 계산하여 관심 대상 객체를 표현하는 방법이 있습니다. HOG feature는 이미지 내 에지 방향의 분포를 인코딩하며, 대상 객체를 나타내기 위해 사용될 수 있습니다. 또 다른 예로는 크기 불변 특징 변환(SIFT)이 있으며, 이미지의 크기, 회전 및 이동에 불변한 지역적 특징점을 감지하고 설명합니다.

* hand-engineered feature는 특정 시나리오에서 학습된 feature보다 일부 이점이 있습니다. 예를 들어, hand-engineered feature는 더 해석 가능하며, 특정 하드웨어 아키텍처에 최적화된 계산이 가능합니다. 그러나 효과적인 hand-engineered feature를 디자인하기 위해서는 상당한 도메인 전문 지식이 필요하며, 문제 도메인의 복잡성과 다양성에 제한될 수 있습니다.

* 딥 러닝 기반 방법의 등장으로 학습된 feature가 hand-engineered feature를 대체하게 되었습니다. 학습된 feature는 보편적이며 수동적인 feature 엔지니어링 없이 다양한 문제 도메인에 적응할 수 있습니다.

Nevertheless, featurizing each level of an image pyra- mid has obvious limitations. Inference time increases con- siderably (e.g., by four times [11]), making this approach impractical for real applications. Moreover, training deep networks end-to-end on an image pyramid is infeasible in terms of memory, and so, if exploited, image pyramids are used only at test time [15, 11, 16, 35], which creates an inconsistency between train/test-time inference. For these reasons, Fast and Faster R-CNN [11, 29] opt to not use fea- turized image pyramids under default settings.


그럼에도 불구하고 image pyramid의 각 level을 featurizing하는 것은 명백한 한계가 있다. inference 시간이 상당히 증가하여(예: [11]의 4배), 이 접근법은 실제 응용에 실용적이지 않다. 또한, image pyramid에서 deep 네트워크를 end-to-end으로 훈련하는 것은 메모리 측면에서 실현 불가능하므로, 악용될 경우 image pyramids는 테스트 시간에만 사용되므로 [15, 11, 16, 35], train/test 시간 inference 간에 불일치가 발생한다. 이러한 이유로, Fast and Faster R-CNN [11, 29]은 기본 설정에서 featurized image pyramids를 사용하지 않기로 선택한다.

However, image pyramids are not the only way to compute a multi-scale feature representation. A deep ConvNet computes a feature hierarchy layer by layer, and with sub- sampling layers the feature hierarchy has an inherent multi- scale, pyramidal shape. This in-network feature hierarchy produces feature maps of different spatial resolutions, but introduces large semantic gaps caused by different depths. The high-resolution maps have low-level features that harm their representational capacity for object recognition.

하지만 image pyramids는 multi scale feature representations을 계산하는 유일한 방법은 아닙니다. deep ConvNet은 layer 별로 feature hierarchy를 계산하고, subsampling layers을 사용하여 feature hierarchy가 내재하는 multi scale, pyramidal shape을 갖습니다. 
이러한 in-network feature hierarchy 구조는 서로 다른 spatial resolutions의 feature maps을 생성하지만 depths의 차이에 의해 발생하는 큰 semantic gaps을 도입합니다. high-resolution maps은 representational 능력을 해치는 low-level features를 가지고 있습니다.


The Single Shot Detector (SSD) [22] is one of the first attempts at using a ConvNet’s pyramidal feature hierarchy as if it were a featurized image pyramid (Fig. 1(c)). Ideally, the SSD-style pyramid would reuse the multi-scale feature maps from different layers computed in the forward pass and thus come free of cost. But to avoid using low-level features SSD foregoes reusing already computed layers and instead builds the pyramid starting from high up in the network (e.g., conv4 3 of VGG nets [36]) and then by adding several new layers. Thus it misses the opportunity to reuse the higher-resolution maps of the feature hierarchy. We show that these are important for detecting small objects.


Single Shot Detector(SSD)[22]는 ConvNet의 pyramidal feature hierarchy을 마치  featurized image pyramid인 것처럼 사용하려는 첫 번째 시도 중 하나입니다(그림 1(c)).  이상적으로 SSD-style pyramid는 forward pass에서 계산된 여러 레이어의 multi-scale feature maps을 재사용하므로 비용이 들지 않습니다.  그러나 low-level features을 사용하지 않기 위해 SSD는 이미 계산된 레이어를 재사용하지 않고 대신 네트워크의 high layer(예: VGG 네트워크의 conv4_3[36])에서 시작하여 여러 새 레이어를 추가하여 pyramids를 구축합니다.  따라서 feature hierarchy의 higher-resolution maps을 재사용할 기회를 놓치게 됩니다.  우리는 이러한 것들이 작은 object를 감지하는 데 중요하다는 것을 보여줍니다.


The goal of this paper is to naturally leverage the pyramidal shape of a ConvNet’s feature hierarchy while creating a feature pyramid that has strong semantics at all scales. To achieve this goal, we rely on an architecture that combines low-resolution, semantically strong features with high-resolution, semantically weak features via a top-down pathway and lateral connections (Fig. 1(d)). The result is a feature pyramid that has rich semantics at all levels and is built quickly from a single input image scale. In other words, we show how to create in-network feature pyramids that can be used to replace featurized image pyramids with- out sacrificing representational power, speed, or memory.

이 논문의 목표는 ConvNet's feature hierarchy의 pyramidal shape을 자연스럽게 활용하면서 모든 scales에서 강력한 semantics을 가진 feature pyramid를 만드는 것입니다.  이 목표를 달성하기 위해 top-down pathway 및 lateral connections을 통해 low-resolution의 semantic적으로 강한 features을 high-resolution의 semantic하게 약한 features과 결합하는 아키텍처에 의존합니다(Fig.1(d)).  그 결과 모든 levels에서 semantics이 풍부하고 단일 입력 이미지 scale에서 신속하게 구축되는 feature pyramid가 생성됩니다.  즉, representational power, 속도 또는 메모리를 희생하지 않고 featurized image pyramids를 대체하는 데 사용할 수 있는 네트워크 in-network feature pyramids를 만드는 방법을 보여줍니다.


Similar architectures adopting top-down and skip connections are popular in recent research [28, 17, 8, 26]. Their goals are to produce a single high-level feature map of a fine resolution on which the predictions are to be made (Fig. 2 top). On the contrary, our method leverages the architecture as a feature pyramid where predictions (e.g., object detections) are independently made on each level (Fig. 2 bottom). Our model echoes a featurized image pyramid, which has not been explored in these works.


top-down 및 skip connections을 채택하는 유사한 아키텍처가 최근 연구[28, 17, 8, 26]에서 인기가 있습니다.  그들의 목표는 예측이 이루어질 fine resolution의 단일 high-level feature map을 생성하는 것입니다(Fig. 2 상단).  반대로, 우리의 방법은 예측(예: object detections)이 각 level에서 독립적으로 수행되는 feature pyramids를 아키텍처로 활용합니다(Fig. 2 하단).  우리의 모델은 이 작업에서 탐구되지 않은 featurized image pyramid pyramids를 반영합니다.

We evaluate our method, called a Feature Pyramid Network (FPN), in various systems for detection and segmentation [11, 29, 27]. Without bells and whistles, we report a state-of-the-art single-model result on the challenging COCO detection benchmark [21] simply based on FPN and a basic Faster R-CNN detector [29], surpassing all existing heavily-engineered single-model entries of competition winners. In ablation experiments, we find that for bounding box proposals, FPN significantly increases the Average Recall (AR) by 8.0 points; for object detection, it improves the COCO-style Average Precision (AP) by 2.3 points and PASCAL-style AP by 3.8 points, over a strong single-scale baseline of Faster R-CNN on ResNets [16]. Our method is also easily extended to mask proposals and improves both instance segmentation AR and speed over state-of-the-art methods that heavily depend on image pyramids.


우리는 detection 및 segmentation을 위한 다양한 시스템에서 Feature Pyramid Network(FPN)라고 불리는 우리의 방법을 평가합니다[11, 29, 27]. 부가 기능 없이, 우리는 FPN과 기본 Faster R-CNN 검출기[29]를 기반으로,  까다로운 COCO detection benchmark[21]에 대한 state-of-the-art single-model result를 보고하여 경쟁 우승자의 기존 모든 엔지니어링 단일 모델 항목을 능가합니다. ablation experiments에서, 우리는 bounding box proposals의 경우, FPN이 Average Recall (AR)을 8.0점 크게 증가시킨다는 것을 발견했습니다. object detection의 경우 ResNets[16]에서 Faster R-CNN의 강력한 single-scale baseline보다 COCO 스타일의 평균 정밀도(AP)를 2.3점, PASCAL 스타일의 AP를 3.8점 향상시킵니다. 우리의 방법은 또한 mask proposals으로 쉽게 확장되며 image pyramids에 크게 의존하는 state-of-the-art 방법에 비해 instance segmentation AR과 속도를 모두 향상시킵니다.

In addition, our pyramid structure can be trained end-to-end with all scales and is used consistently at train/test time, which would be memory-infeasible using image pyramids. As a result, FPNs are able to achieve higher accuracy than all existing state-of-the-art methods. Moreover, this improvement is achieved without increasing testing time over the single-scale baseline. We believe these advances will facilitate future research and applications. Our code will be made publicly available.


또한, 우리의 pyramid 구조는 모든 scales로 end-to-end 훈련이 가능하며, image pyramids를 사용하여 메모리가 불가능할 수 있는 train/test 시간에 일관되게 사용된다.
결과적으로 FPN은 기존의 모든 state-of-the-art 방법보다 더 높은 정확도를 달성할 수 있습니다.  또한 이러한 개선은 single-scale baseline에서 테스트 시간을 늘리지 않고 달성됩니다.  우리는 이러한 발전이 미래의 연구와 응용을 촉진할 것이라고 믿습니다.  우리의 코드는 공개될 것입니다.

# 3. Feature Pyramid Networks


Our goal is to leverage a ConvNet’s pyramidal feature hierarchy, which has semantics from low to high levels, and build a feature pyramid with high-level semantics throughout. The resulting Feature Pyramid Network is general- purpose and in this paper we focus on sliding window pro- posers (Region Proposal Network, RPN for short) [29] and region-based detectors (Fast R-CNN) [11]. We also generalize FPNs to instance segmentation proposals in Sec. 6.
Our method takes a single-scale image of an arbitrary size as input, and outputs proportionally sized feature maps at multiple levels, in a fully convolutional fashion. This pro- cess is independent of the backbone convolutional architectures (e.g., [19, 36, 16]), and in this paper we present results using ResNets [16]. The construction of our pyramid in- volves a bottom-up pathway, a top-down pathway, and lateral connections, as introduced in the following.


우리의 목표는, low levels에서 high levels까지의 semantic을 가진 ConvNet의 pyramidal feature hierarchy을 활용하여 전체적으로 high-level semantics을 가진 feature pyramid를 구축하는 것입니다. 결과적으로 얻어지는 Feature Pyramid Network 는 일반적인 용도로 사용될 수 있으며, 이 논문에서는 sliding window proposers(Region Proposal Network, RPN) [29]과 region-based detectors(Fast R-CNN) [11]에 초점을 맞추고 있습니다. 또한, 우리는 제6절에서 FPN을 instance segmentation proposals으로 일반화시킵니다.
우리의 방법은 임의의 크기의 단일 scale 이미지를 입력으로 사용하고, 전적으로 컨볼루션 방식으로 multiple levels에서 비례적인 크기의 feature maps을 출력합니다. 이 프로세스는 backbone(ConvNet의 핵심 구조) convolutional architectures (예: [19, 36, 16])와는 독립적입니다. 이 논문에서는 ResNet [16]을 사용한 결과를 제시합니다. 우리의 pyramids 구성은 아래에서 소개하는 바와 같이 bottom-up pathway, top-down pathway, 그리고 lateral connections로 이루어져 있습니다.


Bottom-up pathway. The bottom-up pathway is the feed- forward computation of the backbone ConvNet, which computes a feature hierarchy consisting of feature maps at several scales with a scaling step of 2. There are often many layers producing output maps of the same size and we say these layers are in the same network stage. For our feature pyramid, we define one pyramid level for each stage. We choose the output of the last layer of each stage as our reference set of feature maps, which we will enrich to create our pyramid. This choice is natural since the deepest layer of each stage should have the strongest features.


Bottom-up pathway. Bottom-up pathway는 ConvNet backbone의 feed- forward 계산으로 구성되며, 2의 scaling 간격으로 여러 scales의 feature maps을 생성하는 feature hierarchy을 계산합니다. 같은 크기의 output maps을 생성하는 많은 레이어들이 있으며, 이러한 레이어들은 같은 network stage에 있습니다. feature pyramid의 경우 각 stage에 대해 하나의 pyramid level을 정의한다. 우리는 각 stage의 마지막 레이어의 출력을 우리의 feature maps의 참조 세트로 선택하고, 이를 강화하여 pyramid를 만듭니다. 각 stage의 가장 깊은 레이가 가장 강한 features을 가져야 하기 때문에 이러한 선택은 자연스러운 것이다.


Specifically, for ResNets [16] we use the feature activations output by each stage’s last residual block. We denote the output of these last residual blocks as {C2 , C3 , C4 , C5 } for conv2, conv3, conv4, and conv5 outputs, and note that they have strides of {4, 8, 16, 32} pixels with respect to the input image. We do not include conv1 into the pyramid due to its large memory footprint.


특히 ResNet [16]의 경우 각 stage의 마지막 residual block에서 출력되는 feature activations를 사용합니다. 우리는 conv2, conv3, conv4 및 conv5 출력에 대한 이러한 마지막 residual blocks의 출력을 {C2, C3, C4, C5}로 표시하며, 입력 이미지에 대해 {4, 8, 16, 32} 픽셀의 스트라이드를 가진다는 것을 유의해야 합니다. 우리는 큰 메모리 사용량 때문에 conv1을 pyramids에 포함시키지 않습니다.



Top-down pathway and lateral connections. The top-down pathway hallucinates higher resolution features by upsampling spatially coarser, but semantically stronger, feature maps from higher pyramid levels. These features are then enhanced with features from the bottom-up pathway via lateral connections. 

Top-down pathway and lateral connections. top-down pathway는 상위 pyramid levels에서 semantic 강도가 더 높은 공간적으로 더 거친(coarse) feature maps을 upsampling하여 더 높은 resolution features을 생성합니다. 
그런 다음 이러한 features은 lateral connections을 통한 bottom-up pathway의 features으로 향상됩니다.


Each lateral connection merges feature maps of the same spatial size from the bottom-up pathway and the top-down pathway. 


각각의 lateral connection은 bottom-up pathway와 top-down pathway에서 동일한 spatial size의 feature maps을 병합합니다.


The bottom-up feature map is of lower-level semantics, but its activations are more accurately localized as it was subsampled fewer times. {C2, C3, C4, C5} that are respectively of the same spatial sizes.

bottom-up feature map은 상대적으로 lower-level semantics을 가지지만, subsampling 횟수가 적기 때문에 activations이 더 정확하게 localization 됩니다. {C2, C3, C4, C5}는 각각 동일한 spatial sizes를 가집니다.

Fig. 3 shows the building block that constructs our top-down feature maps. With a coarser-resolution feature map, we upsample the spatial resolution by a factor of 2 (using nearest neighbor upsampling for simplicity). The upsampled map is then merged with the corresponding bottom-up map (which undergoes a 1×1 convolutional layer to reduce channel dimensions) by element-wise addition. This process is iterated until the finest resolution map is generated. To start the iteration, we simply attach a 1×1 convolutional layer on C5 to produce the coarsest resolution map. Finally, we append a 3 × 3 convolution on each merged map to generate the final feature map, which is to reduce the aliasing effect of upsampling. This final set of feature maps is called {P2, P3, P4, P5}, corresponding to o {C2, C3, C4, C5} that are respectively of the same spatial sizes.


Fig. 3은 우리의 top-down feature maps을 구성하는 빌딩 블록을 보여줍니다. coarser-resolution feature map과 함께 가장 nearest neighbor upsampling을 사용하여 spatial resolution를 2배로 upsampling 합니다. 이 upsampled map은 해당되는 bottom-up map(채널 차원을 줄이기 위해 1×1 conv layer을 거침)과 element-wise addition으로 병합됩니다. 이 과정은 가장 fine resolution map이 생성될 때까지 반복됩니다. 반복을 시작하기 위해 C5에 1×1 conv 레이어을 단순히 추가하여 가장 coarse resolution map을 생성합니다. 마지막으로, 각  merged map에 3×3 conv을 추가하여 upsampling의 aliasing 효과를 줄인 최종 feature map을 생성합니다. 이 최종 feature maps 세트는 각각 {C2, C3, C4, C5}와 동일한 공간 크기인 {P2, P3, P4, P5}로 불립니다.


Because all levels of the pyramid use shared classifiers/regressors as in a traditional featurized image pyramid, we fix the feature dimension (numbers of channels, denoted as d) in all the feature maps. We set d = 256 in this paper and thus all extra convolutional layers have 256-channel outputs. There are no non-linearities in these extra layers, which we have empirically found to have minor impacts.


전통적인 featurized image pyramid와 마찬가지로, 이 pyramid의 모든 levels에서 공유되어지는 classifiers/regressors를 사용하므로 모든 feature maps에서 feature dimension(channels 수, d로 표시)을 고정합니다. 이 논문에서는 d = 256으로 설정하고 따라서 모든 추가 컨볼루션 레이어는 256채널 출력을 갖습니다. 이러한 추가 레이어에는 비선형성이 없으며, 우리는 경험적으로 그 영향이 미미하다는 것을 발견했습니다.


Simplicity is central to our design and we have found that our model is robust to many design choices. We have exper- imented with more sophisticated blocks (e.g., using multi- layer residual blocks [16] as the connections) and observed marginally better results. Designing better connection mod- ules is not the focus of this paper, so we opt for the simple design described above.

# 4. Applications

Our method is a generic solution for building feature pyramids inside deep ConvNets. In the following we adopt our method in RPN [29] for bounding box proposal generation and in Fast R-CNN [11] for object detection. To demonstrate the simplicity and effectiveness of our method, we make minimal modifications to the original systems of [29, 11] when adapting them to our feature pyramid.

우리의 방법은 깊은 ConvNets 내부에 feature pyramids를 구축하기 위한 일반적인 솔루션입니다.  다음에서 우리는 bounding box proposal을 위한 RPN[29]과 object 감지를 위한 Fast R-CNN[11]의 방법을 채택합니다.  우리 방법의 단순성과 효율성을 입증하기 위해 [29, 11]의 원래 시스템을 feature pyramids에 적용할 때 최소한의 수정을 합니다.


### 4.1. Feature Pyramid Networks for RPN
RPN [29] is a sliding-window class-agnostic object de- tector. In the original RPN design, a small subnetwork is evaluated on dense 3×3 sliding windows, on top of a single- scale convolutional feature map, performing object/non- object binary classification and bounding box regression. This is realized by a 3×3 convolutional layer followed by two sibling 1×1 convolutions for classification and regres- sion, which we refer to as a network head. The object/non- object criterion and bounding box regression target are de- fined with respect to a set of reference boxes called anchors [29]. The anchors are of multiple pre-defined scales and aspect ratios in order to cover objects of different shapes.


RPN[29]은 sliding-window class에 구애받지 않는 object detector입니다.  원래 RPN 디자인에서 소규모 서브네트워크는 object/non-object binary classification 및 bounding box regression를 수행하는 단일 scale convolutional feature map 위에 있는 dense 3×3 sliding windows에서 평가됩니다.  이것은 classification 및 regression를 위한 두 개의 sibling 1×1 conv가 뒤따르는 3×3 conv layer에  의해 실현되며, 우리는 이를 network head라고 합니다.  object/non-object criterion과  bounding box regression target은 anchors 라고 하는 reference boxes세트와 관련하여 정의됩니다[29].  anchor는 서로 다른 모양의 개체를 포함하기 위해 미리 정의된 여러 scales과 aspect ratios입니다.


 We adapt RPN by replacing the single-scale feature map with our FPN. We attach a head of the same design (3×3 conv and two sibling 1×1 convs) to each level on our feature pyramid. Because the head slides densely over all locations in all pyramid levels, it is not necessary to have multi-scale anchors on a specific level. Instead, we assign anchors of a single scale to each level. Formally, we define the anchors to have areas of {322 , 642 , 1282 , 2562 , 5122 } pixels on {P2, P3, P4, P5, P6} respectively. As in [29] we also use anchors of multiple aspect ratios {1:2, 1:1, 2:1} at each level. So in total there are 15 anchors over the pyramid.

single-scale feature map을 FPN으로 대체하여 RPN을 적용합니다.  동일한 디자인의 head(3×3 conv and two sibling 1×1 convs)를 feature pyramid의 각 level에 연결합니다.  head는 모든 pyramid levels의 모든 위치에서 dense하게 slide 하기 때문에 특정 level에 multi scale anchors를 가질 필요가 없습니다.  대신 각 level에 single scale의 anchors를 할당합니다.  공식적으로 {P2, P3, P4, P5, P6}에서 각각 {32^2, 64^2, 128^2, 256^2, 512^2} 픽셀의 영역을 갖도록 anchors를 정의합니다. [29]에서와 같이 각 level에 multiple aspect ratios {1:2, 1:1, 2:1}의 anchors을 사용합니다.  따라서 pyramid에는 총 15개의 anchors가 있습니다.


We assign training labels to the anchors based on their Intersection-over-Union (IoU) ratios with ground-truth bounding boxes as in [29]. Formally, an anchor is assigned a positive label if it has the highest IoU for a given ground- truth box or an IoU over 0.7 with any ground-truth box, and a negative label if it has IoU lower than 0.3 for all ground-truth boxes. Note that scales of ground-truth boxes are not explicitly used to assign them to the levels of the pyramid; instead, ground-truth boxes are associated with anchors, which have been assigned to pyramid levels. As such, we introduce no extra rules in addition to those in [29].
We note that the parameters of the heads are shared across all feature pyramid levels; we have also evaluated the alternative without sharing parameters and observed similar accuracy. The good performance of sharing parameters indicates that all levels of our pyramid share similar semantic levels. This advantage is analogous to that of using a fea turized image pyramid, where a common head classifier can be applied to features computed at any image scale.
With the above adaptations, RPN can be naturally trained and tested with our FPN, in the same fashion as in [29]. We elaborate on the implementation details in the experiments.


우리는 [29]에서와 같이 ground-truth bounding boxes가 있는 IoU(Intersection-over-Union) 비율을 기반으로 anchors에 training labels을 할당합니다.  공식적으로 anchor는 주어진 ground- truth box에 대해 IoU가 가장 높거나 모든 ground- truth box에서 IoU가 0.7 이상인 경우 positive label이 할당되고 모든 ground-truth에 대해 IoU가 0.3보다 낮은 경우 negative label이 지정됩니다.  ground-truth boxes의 scales는 pyramids의 levels에 할당하는 데 명시적으로 사용되지 않습니다.  대신 ground-truth boxes는 pyramids levels에 할당된 anchors와 연결됩니다.  따라서 [29]의 규칙 외에 추가 규칙을 도입하지 않습니다.
 heads의 파라미터들은 모든 feature pyramid levels에서 공유됩니다.  또한 파라미터들을 공유하지 않는 alternative을 평가했으며 유사한 정확도를 관찰했습니다.  공유 파라미터들의 우수한 성능은 pyramids의 모든 levels이 유사한 semantic levels을 공유한다는 것을 나타냅니다.  이 이점은 featurized image pyramid를 사용하는 것과 유사합니다. 여기에서 일반적인 head classifier는 모든 이미지 scale에서 계산된 features에 적용될 수 있습니다.
 위의 적용으로 RPN은 [29]에서와 같은 방식으로 FPN으로 자연스럽게 훈련되고 테스트될 수 있습니다.  우리는 실험에서 구현 세부 사항에 대해 자세히 설명합니다.



## 4.2. Feature Pyramid Networks for Fast R-CNN


Fast R-CNN [11] is a region-based object detector in which Region-of-Interest (RoI) pooling is used to extract features. Fast R-CNN is most commonly performed on a single-scale feature map. To use it with our FPN, we need to assign RoIs of different scales to the pyramid levels.
We view our feature pyramid as if it were produced from an image pyramid. Thus we can adapt the assignment strategy of region-based detectors [15, 11] in the case when they are run on image pyramids. Formally, we assign an RoI of width w and height h (on the input image to the network) to the level Pk of our feature pyramid by: </br>
k = ⌊k0 + log2( wh/224)⌋. (1) </br>
Here 224 is the canonical ImageNet pre-training size, and k0 isthetargetlevelonwhichanRoIwithw×h = 2242 should be mapped into. Analogous to the ResNet-based Faster R-CNN system [16] that uses C4 as the single-scale feature map, we set k0 to 4. Intuitively, Eqn. (1) means that if the RoI’s scale becomes smaller (say, 1/2 of 224), it should be mapped into a finer-resolution level (say, k = 3).


Fast R-CNN[11]은 features 추출에 Region-of-Interest(RoI) pooling을 사용하는 region-based object detector입니다.  Fast R-CNN은 single-scale feature map에서 가장 일반적으로 수행됩니다.  FPN과 함께 사용하려면 pyramid levels에 서로 다른 scales의 RoI를 할당해야 합니다.
 우리는 feature pyramid를 image pyramid에서 생성된 것처럼 봅니다.  따라서 image pyramids에서 실행되는 경우 region-based detectors [15, 11]의 할당 전략을 적용할 수 있습니다.  공식적으로 우리는 다음과 같이 너비 w와 높이 h의 RoI(네트워크에 대한 입력 이미지에서)를 feature pyramids의 수준 Pk에 할당합니다.
  k = ⌊k0 + log2(wh/224)⌋.  (1)
 여기서 224는 표준 ImageNet pre-training 크기이고 k0는 w×h = 224^2인 RoI가 매핑되어야 하는 target level입니다.  단일 scale 기능 맵으로 C4를 사용하는 ResNet 기반 Faster R-CNN 시스템[16]과 유사하게 k0을 4로 설정합니다.  (1)은 RoI의 scale이 작아지면(예: 224의 1/2)  finer-resolution level(예: k = 3)으로 매핑되어야 함을 의미합니다.


We attach predictor heads (in Fast R-CNN the heads are class-specific classifiers and bounding box regressors) to all RoIs of all levels. Again, the heads all share parameters, regardless of their levels. In [16], a ResNet’s conv5 layers (a 9-layer deep subnetwork) are adopted as the head on top of the conv4 features, but our method has already harnessed conv5 to construct the feature pyramid. So unlike [16], we simply adopt RoI pooling to extract 7×7 features, and attach two hidden 1,024-d fully-connected (fc) layers (each followed by ReLU) before the final classification and bounding box regression layers. These layers are randomly initialized, as there are no pre-trained fc layers available in ResNets. Note that compared to the standard conv5 head, our 2-fc MLP head is lighter weight and faster.
Based on these adaptations, we can train and test Fast R- CNN on top of the feature pyramid. Implementation details are given in the experimental section.


우리는 모든 RoI에 대해 predictor heads를 부착합니다. (Fast R-CNN에서 heads는 class-specific classifiers와 bounding box regressors입니다.) levels과 관계없이 모든 heads 는 파라미터들을 공유합니다. [16]에서는 ResNet의 conv5 레이어 (9개의 레이어로 구성된 하위 네트워크)가 conv4 features 위에 올라가는 head로 사용되었지만, 우리의 방법은 이미 conv5를 피처 pyramids를 구성하는 데 사용했습니다. 따라서 [16]과 달리, 우리는 단순히 RoI pooling을 채택하여 7×7 features를 추출하고 최종 classification 및 bounding box regression 레이어 이전에 두 개의 1,024-d fully-connected (fc) 히든 레이어를 부착합니다. 이러한 레이어들은 ResNet에 사전 훈련된 fc 레이어가 없으므로 임의로 초기화됩니다. 표준 conv5 head와 비교하여 우리의 2-fc MLP 헤드는 가볍고 빠릅니다.

이러한 적용을 기반으로 우리는 feature pyramids 위에 Fast R-CNN을 학습하고 테스트할 수 있습니다. 구현 세부 정보는 실험 섹션에서 설명합니다.

It is noteworthy that removing top-down connections (Table 2(d)) significantly degrades the accu- racy, suggesting that Fast R-CNN suffers from using the low-level features at the high-resolution maps.
In Table 2(f), we adopt Fast R-CNN on the single finest scale feature map of P2. Its result (33.4 AP) is marginally worse than that of using all pyramid levels (33.9 AP, Ta- ble 2(c)). We argue that this is because RoI pooling is a warping-like operation, which is less sensitive to the re- gion’s scales. Despite the good accuracy of this variant, it is based on the RPN proposals of {Pk} and has thus already benefited from the pyramid representation.
