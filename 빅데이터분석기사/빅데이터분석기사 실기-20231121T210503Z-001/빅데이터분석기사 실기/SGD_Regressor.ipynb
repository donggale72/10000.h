#SGD_Regressor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
data = pd.read_csv('/content/drive/MyDrive/datasets/insurance.csv')
x = np.array(data['age'])
y = np.array(data['charges'])
x = x.reshape(1338, 1)
y = y.reshape(1338, 1)
sgd_reg = SGDRegressor(max_iter = 1000, random_state = 45)
sgd_reg.fit(x, y.ravel())
```

```python
print('SGD  회귀 모델 결과')
print('절편', sgd_reg.intercept_, '계수', sgd_reg.coef_)
```

```python
print(sgd_reg.score(x, y))
```

```python
x_new = [[19],[64]]
y_hat = sgd_reg.predict(x_new)
print(y_hat)
```

```python
plt.figure(figsize= (10, 5))
plt.plot(x, y, 'b.')
plt.plot(x_new, y_hat, '-r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
