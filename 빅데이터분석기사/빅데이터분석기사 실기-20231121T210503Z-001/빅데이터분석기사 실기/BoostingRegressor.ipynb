#BoostingRegressor

```python
car = pd.read_csv('/content/drive/MyDrive/datasets/CarPrice_Assignment.csv')
car_num = car.select_dtypes(['number'])
features = list(car_num.columns.difference(['car_ID', 'symboling', 'price','Unnamed: 0']))
x = car_num[features]
y = car_num['price']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
print(car.shape, car.info())
car.describe()
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
reg = AdaBoostRegressor()
pred = reg.fit(x_train, y_train).predict(x_test)

mse = mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
rmse = np.sqrt(mse)
r2= r2_score(y_test, pred)
acc = reg.score(x_test, y_test) #위치 주의  회귀문제에도 acc 쓸수 있음

print('mse : {}'.format(round(mse, 2)))
print('mae : {}'.format(round(mae, 2)))
print('rmse : {}'.format(round(rmse, 2)))
print('r2 : {}'.format(round(r2, 2)))
print('정확도 : {}%'.format(round(acc *100, 2)))
```

```python
importances = reg.feature_importances_
columns_nm= pd.DataFrame(features)
feature_importances = pd.concat([columns_nm, pd.DataFrame(importances)], axis = 1)

feature_importances.columns = ['feature_nm', 'impotances']
print(feature_importances.sort_values(by='impotances', ascending = False))
```
