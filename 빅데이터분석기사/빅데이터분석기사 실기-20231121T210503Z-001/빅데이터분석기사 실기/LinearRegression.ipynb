```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/drive/MyDrive/datasets/insurance.csv')
```

```python
data.info()
```

```python
x = data['age']
y = data['charges']
plt.figure(figsize = (10, 5))
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array(data['age'])
y = np.array(data['charges'])
x = x.reshape(1338, 1) # fit 에 입력되는 데이터의 shape을 2차원으로 변경해야 됨
y = y.reshape(1338, 1)
lr = LinearRegression()
lr.fit(x, y)
```

```python
print('선형 회귀 모델 결과')
print('절편', lr.intercept_, '계수', lr.coef_)
```


Returns
    score : float
        :math:`R^2` of ``self.predict(X)`` w.r.t. `y`.

```python
print(lr.score(x, y)) # lr.score로 결정계수를 바로 구할 수 있음
```

```python
x_new = [[19],[64]]
y_hat = lr.predict(x_new)
print(y_hat)
```

```python
plt.figure(figsize= (10, 5))
plt.plot(x, y, 'b.')
plt.plot(x_new, y_hat, '-r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
