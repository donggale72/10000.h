```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 예제 데이터 생성
group1 = np.random.normal(loc=5, scale=1, size=100)
group2 = np.random.normal(loc=7, scale=1, size=100)
group3 = np.random.normal(loc=6, scale=1, size=100)

# t-test 수행
t_statistic, p_value, _ = sm.stats.ttest_ind(group1, group2)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# 일원분산분석(One-Way ANOVA) 수행
data = np.concatenate([group1, group2, group3])
group_labels = np.array(['Group 1'] * 100 + ['Group 2'] * 100 + ['Group 3'] * 100)
model = ols('data ~ group_labels', data=pd.DataFrame({'data': data, 'group_labels': group_labels}))
anova_result = sm.stats.anova_lm(model.fit())
print(anova_result)
```
