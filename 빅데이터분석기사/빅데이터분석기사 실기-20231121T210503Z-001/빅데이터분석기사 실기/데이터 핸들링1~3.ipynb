```python
import pandas as pd
score = pd.DataFrame({'국어':[100, 80], '수학': [75, 90], '영어': [75, 98]}, index = ['흥부', '놀부'])
score.loc['흥부', '영어'] =100
score['국어'] = score['국어'] - 10
score
```

```python
students = pd.DataFrame({'국어':[ 70, 55], '수학': [77, 90], '영어':[95, 65]}, index = ['콩지', '팥지'])
score = score.append(students)
new_students = pd.Series({'국어': 85, '수학': 55, '영어': 95}, name = '제비')
score = score.append(new_students)
score
```

```python
score2= score
new_students2 = {'국어': 68, '수학': 77, '영어': 88} #Dictionary 로 행을 추가시 오류발생함
score2.append(new_students2, ignore_index = True)
```

```python
science = [80, 70, 90, 75, 60]
score['과학'] = science
score['학년'] = 1
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['과학'] + score['영어']
score
```

```python
score2=score
score.drop('흥부', inplace = True)
score.drop(columns =['과학', '학년', '총점'], inplace = True)
score
```

```python
score2.drop('놀부', inplace = False)
score2.drop(columns =['수학'], inplace = False)
score2
#inplace 인수가 False로 설정된 경우, drop() 메서드는 원본 데이터 프레임을 변경하지 않고 새로운 데이터 프레임을 반환
```
