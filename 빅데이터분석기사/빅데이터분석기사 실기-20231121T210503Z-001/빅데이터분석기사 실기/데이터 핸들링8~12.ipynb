##데이터 재구조화

```python
import pandas as pd
import numpy as np
score = { '학년':[1, 1,1,1,2,2], '반':['A', 'A', 'B', 'B','C', 'C'], '성별': [ '여자','남자', '여자','남자', '여자','남자'], '성적':[76, 88, 85, 72, 68, 70]}
score = pd.DataFrame(score)
score
```

```python
# score.apply(np.sqrt, axis = 0)  #apply 함수는 열의 각각의 값에 접근하는데 열의 성별이 str이라서 에러가 발생함
```

```python
score = score.pivot_table(index = ['학년', '반'], columns='성별', values = '성적') # 데이터 재구조화
score
```

```python
score.reset_index().melt(id_vars = ['학년','반'], var_name ='성별', value_name = '성적')  # 데이터 재구조화
```

```python
import pandas as pd
import numpy as np
score.apply(np.sqrt, axis = 0) #np.square() 제곱함수
# melt() 함수를 적용한 후에는 DataFrame의 구조가 피벗테이블과 유사하게 취급됨. 따라서, apply() 함수를 사용하여 피벗테이블 형태를 유지하면서 특정 열에 함수를 적용할 수 있습
```

```python
score.apply(np.max, axis = 0)
```

```python
def plus_five(val):
  return val + 5

score.apply(plus_five)
```

##문자열 데이터 변환

```python
import pandas as pd
landmark = pd.DataFrame({'name':['광화문','호미곳','첨성대'], 'location': ['서울 종로구 사직로 161', '경북 포항시 남구 호미곳면 대보리 150','경북 경주시 인왕동 839-1']})
landmark['location'].str[3:6]
```

##문자열 분할
