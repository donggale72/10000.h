###결측값 처리


```python
import pandas as pd
students = pd.DataFrame({'이름':['장화','홍련','콩쥐','팥지', '해님', '달님'], '국어':[70, 60, None, 100, None, 85], '수학': [65, 100, 80, 95, None, 75]})
students[students['이름'] == '장화']
```

```python
students[(students['국어'] >=80) & (students['수학'] >=80)]
```

```python
students.loc[6, '이름' : '수학'] = ['별님', 50, 60]
```

```python
students.loc[(students['국어'] >=70)&(students['수학']>=60), '합격'] = 'pass' #칼럼명이 존재하지 않으면 ('합격')새롭게 생성
students.loc[students['합격']!='pass', '합격'] = 'Fail'
students
```

```python
import numpy as np
condition_list = [(students['국어']>= 90), (students['국어']>= 80) & (students['국어']<90), (students['국어']>=70)&(students['국어']<80)]
choice_list = ['A','B', 'C']
students['점수'] = np.select(condition_list,choice_list, default='F') #여러개의 조건에 따라 다른 값을 반환하는 경우np.select(조건목록, 선택목록 default=값)
students
```

```python
students.isna()
```

```python
students.notna()
```

```python
students.isna().sum() #열의 합계
```

```python
students.isna().sum(1) #행의 합계
```

```python
students.dropna(thresh=4) #결측값이 아닌 값이 4보다 많은 행만 남기기 5행 해님 삭제됨
```

##정렬


```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
iris = pd.DataFrame(iris.data, columns = iris.feature_names)
iris.sort_index(ascending = False, inplace = True) #ascending =  True (0 부터 오름차순)이면 0부터 시작 ,False 이면 끝에서 부터 (내림차순) 시작 (올라가면 적어지고, 내려가면 적어짐)
iris.head()
```

```python
iris.sort_index(axis = 0, ascending = False, inplace = True) # 1 이면 컬럼명을 기준으로 정렬, 0이면 인덱스를 기준으로 정렬
iris.head()   # 주의 axis 0 일때 인덱스 오름차순으로 정렬되지만 1일 경우 인덱스가 내림차순으로 정렬됨
```

```python
iris.sort_values('petal length (cm)', axis = 0 , ascending = True, inplace = True)
iris
```

```python
iris.sort_values(['petal length (cm)', 'sepal length (cm)'], axis = 0 , ascending = True, inplace = True)
iris  #14, 35 --> 35, 14 변경
```

###데이터 결합


```python
import pandas as pd
HR1 = pd.DataFrame({'이름':['장화','홍련'], '부서': ['영업', '회계'], '직급': [ '팀장', '사원']})
HR2 = pd.DataFrame({'이름':['콩쥐','팥쥐'], '직급': ['사원', '팀장'], '부서': [ '영업', '인사']})
pd.concat([HR1, HR2], axis = 0)
```

```python
pd.concat([HR1, HR2], axis = 0, ignore_index = True) # 기준 인덱스 무시하고 0부터 시작
```

```python
HR3 = pd.DataFrame({'이름':['콩쥐','팥쥐'], '부서': [ '영업', '인사'], '급여':[3500, 2800]}) # 데이터 프레임은 칼럼이 다른 경우 모든 칼럼이 출력
pd.concat([HR1,HR3], axis = 0, ignore_index = True)
```

```python
HR4 = pd.Series({1: 2500}, name = '급여') #Series로 딕서너리의 위치에 값을 대입
pd.concat([HR1, HR4], axis = 1)
```

```python
HR5 = pd.DataFrame({'급여':[4500, 3000, 3500]}) #정보가 없는 칸은 Nan으로 채움
pd.concat([HR1, HR5], axis = 1)
```

###데이터 집계

```python
import numpy as np
import pandas as pd
IRIS = load_iris()
iris = pd.DataFrame(IRIS.data, columns = IRIS.feature_names)
iris['class']=IRIS.target
iris['class'] = iris['class'].map({0:'setosa',1:'vergicolor',2:'virginica'}) # map(함수, 배열) 함수없이 딕셔너리로 넣으면 키값에 요소들이 적용됨
iris.groupby(by = 'class').mean() # 범주형 자료 기술통계
```

```python
iris.describe()
```

```python
iris['class'].describe()# 범주형 자료로 되어 있음
```

```python
iris.groupby(by = 'class').median()
```

```python
pd.Series(iris['class']).value_counts()
```

```python
iris['petal width level'] = pd.qcut(iris['petal width (cm)'], q = 3, labels = ['short', 'middle', 'long']) # pd.qcut() 함수는 데이터프레임의 열을 분위수로 나누는 함수
#data: 분위수를 계산할 열
#q: 분위수의 개수
#labels: 분위수에 대한 레이블
#iris['petal width level']
pd.Series(iris['petal width level']).value_counts().to_frame()
#Pandas의 .to_frame() 메서드는 Series를 DataFrame으로 변환하는 메서드
```

```python
pd.crosstab(iris['petal width level'], iris['class']) #(index, columns, dropna(na를 포함할지 여부 기본 False), nomalize(정규화))
```
