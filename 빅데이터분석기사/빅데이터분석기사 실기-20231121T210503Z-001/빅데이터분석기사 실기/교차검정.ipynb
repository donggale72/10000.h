#적합성검정   (관측빈도의 분포가 기대빈도의 분포를 따르는지 검정)

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/titanic.csv')
df.to_csv('/content/drive/MyDrive/datasets/titanic.csv')
```

```python
df.head()
```

```python
df_t = df[df['survived']==1]
table = df_t[['sex']].value_counts() # 남성과 여성으로 값을 계산해서 분할
df_t.head()
```

```python
table
```

```python
from scipy.stats import chisquare
chi = chisquare(table, f_exp = [171, 171]) #기대빈도 (233+109)/2 ---> 171 적합성 검정은 관측빈도의 분포가 기대빈도의 분포를 따르는지 검정 예) 주사위의 값 확률은 1/6인지 확인
print('<적합도검정>\n', chi) # 남자와 여자의 비율이 같지 않다. 연구가설 채택
```

# 독립성 검정 (두집단간에 연관성이 있는지 검정)

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/titanic.csv')
df.head()
```

```python
table = pd.crosstab(df['class'], df['survived']) # 교차표는 범주형 데이터를 분리할 때에 유용함
table
```

```python
from scipy.stats import chi2_contingency
chi, p, df, expect = chi2_contingency(table)
print('Statistic:', chi)
print('p_value :', p)
print('df :', df)
print('expect: \n', expect) # 두변수간 연관성이 있다. 연구가설 채택 p_value : 4.549251711298793e-23
```

# 동질성 검정  (여러 집단간 분포가 동질적인지 검정)

```python
import scipy.stats as stats
import numpy as np

# 예제 데이터셋: 지역별 정치적 경향에 대한 투표 수
# 행은 다른 지역을 나타내고, 열은 정치적 경향(보수, 중도, 진보)를 나타냄.
data = np.array([[30, 60, 10],  # 지역 A
                 [40, 50, 10],  # 지역 B
                 [30, 40, 30]]) # 지역 C

# 카이제곱 동질성 검정 수행
chi2, p, dof, expected = stats.chi2_contingency(data)

print(f"Chi-squared: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{expected}") #p-value: 0.000200420409482948 분포가 동질적이지 않으므로 연구가설 '지역 간에 정치적 경향의 분포가 다르다' 채택
```
